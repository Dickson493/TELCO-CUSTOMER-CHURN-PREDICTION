```python
# To call Basic Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('telecom_churn.csv')
df
```

```python
#To check the null values
df.isnull().sum()
```

```python
#To get the basic info about the dataseta nd to check for object type variable
df.info()
```

```python
#Total Charges has all numerical value dtype showing object
df['TotalCharges'].value_counts()
#There are 11 special character(space)
```

```python
#Replacing the space with the null value
df['TotalCharges'].replace(' ',np.nan,inplace=True)
```

```python
#changing the data type from object to float
df['TotalCharges']=df['TotalCharges'].astype(float)
```

```python
#Tp check if sucessfully coverted or no
df['TotalCharges'].dtype
```

```python
df['TotalCharges'].fillna(df['TotalCharges'].mean(),inplace=True)
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df_cat=df.select_dtypes(object)
df_num=df.select_dtypes(['float','int'])
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
for i in df_cat:
    df_cat[i]=le.fit_transform(df_cat[i])
```

```python
df_cat
```

```python
df_new=pd.concat([df_num,df_cat],axis=1)
```

```python
df_new
```

```python
df_new.info()
```

```python
#Seperating the input and output column
X=df_new.drop('Churn',axis=1)
Y=df_new['Churn']
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
#train test split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=1)
```

```python
#creating the object of the class
ss=StandardScaler()
```

```python
#performing standard scaler on the input column
X_train=ss.fit_transform(X_train)
X_test=ss.transform(X_test)
```

```python
#checking if the data of the output column(churn) is balanced or no
sns.countplot(data=df_new,x='Churn')
c=df_new['Churn'].value_counts()
plt.yticks(c)
plt.show()
```

```python
#Here we see that the column is imbalance hence we need to balance the data
#Now we perform sampling technique
```

```python
#Performing Over sampling
from imblearn.over_sampling import RandomOverSampler
```

```python
ros=RandomOverSampler()
X_sample1_train,Y_sample1_train=ros.fit_resample(X_train,Y_train)
print('training')
print(pd.Series(Y_train).value_counts())
print(pd.Series(Y_sample1_train).value_counts())

X_sample1_test,Y_sample1_test=ros.fit_resample(X_test,Y_test)
print('teting')
print(pd.Series(Y_test).value_counts())
print(pd.Series(Y_sample1_test).value_counts())
```

```python
#perform random under sampler
```

```python
from imblearn.under_sampling import RandomUnderSampler
```

```python
rus=RandomUnderSampler()
X_sample2_train,Y_sample2_train=rus.fit_resample(X_train,Y_train)
print(pd.Series(Y_train).value_counts())
print(pd.Series(Y_sample2_train).value_counts())
X_sample2_test,Y_sample2_test=rus.fit_resample(X_test,Y_test)
print(pd.Series(Y_test).value_counts())
print(pd.Series(Y_sample2_test).value_counts())
```

```python
#Nowe we have Done Sampling
#Now we will check which one is better by performing base line model(Logistic Regression)
```

```python
#creating a function with over sampling
def create_model1(model):
    model.fit(X_sample1_train,Y_sample1_train)
    Y_pred=model.predict(X_sample1_test)
    print(classification_report(Y_sample1_test,Y_pred))
    print(confusion_matrix(Y_sample1_test,Y_pred))
    return model
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
#first checking lr with oversampling
create_model1(lr)
```

```python
#creating a function with under sampling
def create_model2(model):
    model.fit(X_sample2_train,Y_sample2_train)
    Y_pred=model.predict(X_sample2_test)
    print(classification_report(Y_sample2_test,Y_pred))
    print(confusion_matrix(Y_sample2_test,Y_pred))
    return model
```

```python
lr=create_model2(lr)
```

```python
#Here we understood that Oversampling is better over here
```
